```python
import shutil
for cls in os.listdir(pet_images_folder):
    os.mkdir(os.path.join(reduced_data_folder,cls))
    for image in os.listdir(os.path.join(pet_images_folder,cls))[:number_of_images]:
        print(image)
        shutil.copy(os.path.join(pet_images_folder,cls,image),os.path.join(reduced_data_folder,cls,image))
```
